```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    "../data/processed/creator_feature_engineered.csv"
)

sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="creator_tier"
)

plt.title("Creator Tier Distribution")
plt.xlabel("Creator Tier")
plt.ylabel("Number of Posts")

plt.tight_layout()

plt.savefig(
    "../visuals/creator_tier_distribution.png"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="media_type",
    y="engagement_rate"
)

plt.title("Engagement Rate by Media Type")

plt.tight_layout()

plt.savefig(
    "../visuals/media_type_performance.png"
)

plt.show()
```

```python
category_eng = (
    df.groupby("content_category")
    ["engagement_rate"]
    .mean()
    .sort_values(ascending=False)
)

category_eng.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title(
    "Average Engagement by Content Category"
)

plt.ylabel("Engagement Rate")

plt.tight_layout()

plt.savefig(
    "../visuals/content_category_performance.png"
)

plt.show()
```

```python
traffic = (
    df.groupby("traffic_source")
    ["engagement_rate"]
    .mean()
    .sort_values(ascending=False)
)

traffic.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title(
    "Average Engagement by Traffic Source"
)

plt.tight_layout()

plt.savefig(
    "../visuals/traffic_source_analysis.png"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="has_call_to_action",
    y="engagement_rate"
)

plt.title(
    "CTA Impact on Engagement"
)

plt.tight_layout()

plt.savefig(
    "../visuals/cta_impact.png"
)

plt.show()
```

```python
hourly = (
    df.groupby("post_hour")
    ["engagement_rate"]
    .mean()
)

plt.figure(figsize=(12,5))

hourly.plot(
    marker="o"
)

plt.title(
    "Average Engagement by Posting Hour"
)

plt.xlabel("Hour")
plt.ylabel("Engagement Rate")

plt.tight_layout()

plt.savefig(
    "../visuals/posting_hour_analysis.png"
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=df,
    x="day_of_week",
    y="engagement_rate"
)

plt.xticks(rotation=45)

plt.title(
    "Engagement by Day of Week"
)

plt.tight_layout()

plt.savefig(
    "../visuals/day_of_week_analysis.png"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="reach",
    y="engagement_rate"
)

plt.title(
    "Reach vs Engagement"
)

plt.tight_layout()

plt.savefig(
    "../visuals/reach_vs_engagement.png"
)

plt.show()
```

```python
plt.figure(figsize=(12,8))

corr_cols = [
    "likes",
    "comments",
    "shares",
    "saves",
    "reach",
    "impressions",
    "engagement_rate",
    "reach_rate",
    "save_rate",
    "virality_score",
    "performance_score"
]

sns.heatmap(
    df[corr_cols].corr(),
    annot=True
)

plt.title(
    "Correlation Heatmap"
)

plt.tight_layout()

plt.savefig(
    "../visuals/correlation_heatmap.png"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="performance_category"
)

plt.title(
    "Performance Category Distribution"
)

plt.tight_layout()

plt.savefig(
    "../visuals/performance_category_distribution.png"
)

plt.show()
```
